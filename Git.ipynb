## Git, GitHub, Bash, Commandline etc

### Command Line Commands
- **pwd** -> show me in which directory I am
- **cd** /User/josephinebiedermann/Desktop/... -> move around directories on your machine
- **cd ..** -> goes up a directory
- **cd ~** -> goes to home user
- **ls** -> lists all content of directory
- **ls -a** -> show hidden content of directory
- **ls -la** -> show hidden content of directory (if "xxx.git" is in there, the folder is a git folder)
- **touch** -> create new document in current directory
- **open** -> opens the declared file
- **mkdir** -> create a new folder/directory
- **mv** -> move a file from directory to another mv "file.extention" /User/josephinebiedermann
- **echo** -> run variables in a document echo 'text/variable' > "file.extention"
- **jupyter notebook** -> opens new notebook in current directory
- **rmdr** -> remove directory
- **history** -> shows all you done before


### Git Commands

- git clone 'link from GitHub repository'
- git status
- git add .
- git add "filename.extention"
- git commit -m "comment/message"
- git commit -am "comment/message" -> adds and commits at the same time only worked for modified files, that have been pushed before
- git push
- git pull
- git reset commit_hash -> undo commits, commit hash get with git log
- git log -> shows all the commits made
- git reset -- hard -> delte the commits


#### Pull Request (PR)
- git pr 


#### branches
- git branch -> shows you which branches there are *is the one you are at at the moment
- git checkout existing.branchname -> to switch branches
- git checkout -b new.branchname -> to create new branch
- git diff -> compared version of code
- git merge -> merge locally branches
- git branch -d branchname -> delete desired branch
